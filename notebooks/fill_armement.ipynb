```python
import pandas as pd
from tqdm import tqdm

file_path = '_SELECT_r_p_v_vd_FROM_record_r_JOIN_person_p_ON_r_person_index_p_202408261231.csv'
df = pd.read_csv(file_path).drop('Unnamed: 0', axis=1)
file_path = 'combined.csv'
df1 = pd.read_csv(file_path).drop('Unnamed: 0', axis=1)
```

```python
for index, row in df1.iterrows():
    if pd.notna(row['Remarks']):
        if isinstance(row['Disemb_loc'], str) and 'armement' in row['Disemb_loc'].lower():
            loc = list(df[
                ((df['last_name'] == row['Last Name']) | (df['last_name'].isna() & pd.isna(row['Last Name']))) &
                ((df['first_name'] == row['First Name']) | (df['first_name'].isna() & pd.isna(row['First Name']))) &
                ((df['emb_date'] == row['Emb_date']) | (df['emb_date'].isna() & pd.isna(row['Emb_date'])))
            ]['steps'])
            if len(loc) > 0:
                df1.at[index, 'Disemb_loc'] = loc[0].split("-")[-1]

        if isinstance(row['Emb_loc'], str) and 'armement' in row['Emb_loc'].lower():
            loc = list(df[
                ((df['last_name'] == row['Last Name']) | (df['last_name'].isna() & pd.isna(row['Last Name']))) &
                ((df['first_name'] == row['First Name']) | (df['first_name'].isna() & pd.isna(row['First Name']))) &
                ((df['emb_date'] == row['Emb_date']) | (df['emb_date'].isna() & pd.isna(row['Emb_date'])))
            ]['steps'])
            if len(loc) > 0:
                df1.at[index, 'Emb_loc'] = loc[0].split(" ")[0]
```

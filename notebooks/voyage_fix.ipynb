```python
import pandas as pd
```

```python
df1 = pd.read_csv('voypp1.csv', low_memory=False)[['record_id', 'voyage_id [old]', 'voyage_id [new]', 'document_ID [new]', 'voyage_code CORR']].rename(columns={'document_ID [new]': 'document_id', 'voyage_code CORR': 'voyage_code'})
df2 = pd.read_csv('voypp2.csv', low_memory=False)[['record_id', 'voyage_id [old]', 'voyage_id [new]', 'document_ID [new]', 'voyage_code CORR']].rename(columns={'document_ID [new]': 'document_id', 'voyage_code CORR': 'voyage_code'})
df2 = df2[df2['record_id']!='record_id']
df = pd.concat([df1, df2])
df_v = pd.read_csv('voyage_final.csv', low_memory=False).drop('Unnamed: 0', axis=1)
df_r = pd.read_csv('record_final.csv', low_memory=False).drop('Unnamed: 0', axis=1)
df_r_columns = list(df_r.columns)
t1 = pd.read_csv('voypp1.csv', low_memory=False).rename(columns={'voyage_id [new]': 'voyage_id', 'document_ID [new]': 'document_id'})[df_r_columns]
t2 = pd.read_csv('voypp2.csv', low_memory=False).rename(columns={'voyage_id [new]': 'voyage_id', 'document_ID [new]': 'document_id'})[df_r_columns]
t2 = t2[t2['record_id']!='record_id']
t = pd.concat([t1, t2])
```

```python
r1 = list(t['record_id'])
r2 = list(df_r['record_id'])
over = []
for r in r1:
    if r in r2:
        over.append(r)
```

```python
chunk = t[~t['record_id'].isin(over)]
```

```python
test = pd.concat([df_r, chunk])
```

```python
errors = []
a = list(test['record_id'])
for i in range(len(test)):
    if i not in a:
        errors.append(i)
```

```python
import pandas as pd
from fuzzywuzzy import fuzz
import numpy as np
from tqdm import tqdm
import pickle
```

```python
file_path = 'MOTHER_TABLE.csv'
df = pd.read_csv(file_path).drop('Unnamed: 0', axis=1)
df
```

```python
def consolidate_matches(*rules_matches):
    all_matches = []
    print("Running Consolidate Matches")
    for matches in rules_matches:
        all_matches.extend(matches)
    
    match_dict = {}
    id_counter = 1

    for match in tqdm(all_matches):
        idx1, idx2, certainty = match
        if idx1 not in match_dict and idx2 not in match_dict:
            match_dict[idx1] = (id_counter, certainty)
            match_dict[idx2] = (id_counter, certainty)
            id_counter += 1
        elif idx1 in match_dict and idx2 not in match_dict:
            match_dict[idx2] = (match_dict[idx1][0], certainty)
        elif idx2 in match_dict and idx1 not in match_dict:
            match_dict[idx1] = (match_dict[idx2][0], certainty)
            

    df['Matched_ID'] = df.index.map(lambda idx: match_dict[idx][0] if idx in match_dict else np.nan)
    df['Certainty'] = df.index.map(lambda idx: match_dict[idx][1] if idx in match_dict else 0)
    return df
```

```python
result_df = consolidate_matches(rule_0_matches, rule_1_matches, rule_2_matches, rule_3_matches, rule_4_matches)

print(f"Total matches found: {result_df['Matched_ID'].notna().sum()}")

result_df.to_csv('Matched_Results.csv', index=False)
```

```python
import pandas as pd
from tqdm import tqdm
```

```python
df_p = pd.read_csv('person_by_id.csv', low_memory=False).drop('Unnamed: 0', axis=1)
df_p
```

```python
ous = list(df_p[df_p['alternate_first_name'].notna() & df_p['alternate_first_name'].str.contains('ou')]['person_id'])
```

```python
len(ous)
```

```python
for index, row in tqdm(df_p.iterrows()):
    if pd.notna(row['alternate_first_name']):
        al_fn = row['alternate_first_name']
        if isinstance(al_fn, str):
            temp = al_fn.split(" ")
            if len(temp) > 1 and temp[0]=='ou':
                df_p.at[index, 'alternate_first_name'] = " ".join(temp[1:])
    if pd.notna(row['alternate_last_name']):
        al_ln = row['alternate_last_name']
        if isinstance(al_ln, str):
            temp = al_ln.split(" ")
            if len(temp) > 1 and temp[0]=='ou':
                df_p.at[index, 'alternate_last_name'] = " ".join(temp[1:])
```

```python
df_p.to_csv('person_by_id.csv')
```

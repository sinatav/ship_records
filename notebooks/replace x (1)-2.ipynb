```python
def percentile_classification(lst):
    temp = []
    column = lst
    d_value = []
    for i in column:
        if i not in d_value:
            d_value.append(i)
    for i in column:
        if i < np.nanpercentile(d_value, 20):
            temp.append(0)
        elif np.nanpercentile(d_value, 20) <= i < np.nanpercentile(d_value, 40):
            temp.append(1)
        elif np.nanpercentile(d_value, 40) <= i < np.nanpercentile(d_value, 60):
            temp.append(2)
        elif np.nanpercentile(d_value, 60) <= i < np.nanpercentile(d_value, 80):
            temp.append(3)
        elif np.nanpercentile(d_value, 80) <= i:
            temp.append(4)
    return temp
```

```python
def average_lists(lists):
    # Determine the length of the lists
    list_length = len(lists[0])
    
    # Initialize an empty list to store averages
    averages = []
    
    # Iterate through each position in the lists
    for i in range(list_length):
        total = 0
        # Iterate through each list and sum up the values at the current position
        for lst in lists:
            total += lst[i]
        # Calculate the average for the current position and append it to the averages list
        averages.append(total / len(lists))
    
    return averages
```

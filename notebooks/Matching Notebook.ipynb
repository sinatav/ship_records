```python
import pandas as pd
from fuzzywuzzy import fuzz
import numpy as np
from tqdm import tqdm
import pickle

def assign_match_id(dfr, rule):
    index = 1
    for i in tqdm(list(rule.keys())):
        assign = 0
        if dfr.at[i, 'Match_id'] == 0:
            for j in rule[i]:
                if dfr.at[j, 'Match_id'] != 0:
                    assign = dfr.at[j, 'Match_id']
                    break
            if assign != 0:
                dfr.at[i, 'Match_id'] = assign
                for j in rule[i]:
                    dfr.at[j, 'Match_id'] = assign
            else:
                dfr.at[i, 'Match_id'] = index
                for j in rule[i]:
                    dfr.at[j, 'Match_id'] = index
                index += 1
                
file_path = 'MOTHER_TABLE.csv'
df = pd.read_csv(file_path).drop('Unnamed: 0', axis=1)

with open('rule_0_matches.pickle', 'rb') as f:
    rule_0_matches = pickle.load(f)
with open('rule_1_v2.pickle', 'rb') as f:
    rule_1_matches = pickle.load(f)
with open('rule_2_matches.pickle', 'rb') as f:
    rule_2_matches = pickle.load(f)
with open('rule_3_matches.pickle', 'rb') as f:
    rule_3_matches = pickle.load(f)
with open('rule_4_matches.pickle', 'rb') as f:
    rule_4_matches = pickle.load(f)
    
rule_0 = {}
for i in tqdm(rule_0_matches):
    if i[2] > 5.5 and i[0] != i[1]:
        if i[0] in (rule_0.keys()):
            rule_0[i[0]].append(i[1])
        else:
            rule_0[i[0]] = [i[1]]
rule_1 = {}
for i in tqdm(rule_1_matches):
    if i[0] != i[1]:
        if i[0] in (rule_1.keys()):
            rule_1[i[0]].append(i[1])
        else:
            rule_1[i[0]] = [i[1]]
rule_2 = {}
for i in tqdm(rule_2_matches):
    if i[0] != i[1]:
        if i[0] in (rule_2.keys()):
            rule_2[i[0]].append(i[1])
        else:
            rule_2[i[0]] = [i[1]]
rule_3 = {}
for i in tqdm(rule_3_matches):
    if i[0] != i[1]:
        if i[0] in (rule_3.keys()):
            rule_3[i[0]].append(i[1])
        else:
            rule_3[i[0]] = [i[1]]
rule_4 = {}
for i in tqdm(rule_4_matches):
    if i[0] != i[1]:
        if i[0] in (rule_4.keys()):
            rule_4[i[0]].append(i[1])
        else:
            rule_4[i[0]] = [i[1]]
            
df_0 = df.copy()
df_1 = df.copy()
df_2 = df.copy()
df_3 = df.copy()
df_4 = df.copy()

df_0['Match_id'] = 0
df_1['Match_id'] = 0
df_2['Match_id'] = 0
df_3['Match_id'] = 0
df_4['Match_id'] = 0

assign_match_id(df_0, rule_0)
assign_match_id(df_1, rule_1)
assign_match_id(df_2, rule_2)
assign_match_id(df_3, rule_3)
assign_match_id(df_4, rule_4)
```

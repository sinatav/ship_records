```python
import pandas as pd
```

```python
df_r = pd.read_csv('record_correcte.csv', low_memory=False).rename(columns={'id': 'record_id'})
```

```python
index_cols = ['person_index', 'origin_index', 'function_index', 'ship_index', 'voyage_index', 'record_index']
df = df_r.copy()
```

```python
for col in index_cols:
    df[col] = df[col].astype(str)
    df = df[df[col].notna()]
    df[col] = df[col].str.split('/')
    df = df.explode(col)
    df[col] = df[col].astype(int)
```

```python
df.to_csv('record_exploded.csv')
```

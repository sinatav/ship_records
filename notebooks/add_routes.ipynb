```python
def merge_consecutive_same_values(df, name, operation, date, voyage_code, note, ship):
    merged_df = pd.DataFrame(columns=df.columns)
    current_row = None
    start_date = None
    end_date = None
    current_voyage_code = None
    stop_id = 0
    route_id_dict = {}
    route_id_counter = 1
    
    for index, row in df.iterrows():
        if current_row is None:
            current_row = row
            start_date = row[date]
            end_date = row[date]
            current_voyage_code = row[voyage_code]
            stop_id = 1
            if current_voyage_code not in route_id_dict:
                route_id_dict[current_voyage_code] = route_id_counter
                route_id_counter += 1
        elif row[name] == current_row[name] and row[voyage_code] == current_row[voyage_code]:
            current_row = current_row.combine_first(row)
            end_date = row[date]
            current_row[date] = f"{start_date} to {end_date}"
            if pd.notnull(row[operation]):
                current_row[operation] = ', '.join([str(val) for val in set(current_row[operation].split(', ') + row[operation].split(', '))])
            if pd.notnull(row[note]):
                current_row[note] = ', '.join([str(val) for val in set(current_row[note].split(', ') + row[note].split(', ')) if pd.notnull(val)])
            current_row[ship] = ', '.join([str(val) for val in set(current_row[ship].split(', ') + row[ship].split(', '))])
        else:
            current_row['stop_id'] = stop_id
            current_row['route_id'] = route_id_dict[current_voyage_code]
            merged_df = pd.concat([merged_df, current_row.to_frame().transpose()], ignore_index=True)
            current_row = row
            start_date = row[date]
            end_date = row[date]
            current_voyage_code = row[voyage_code]
            stop_id = 1
            if current_voyage_code not in route_id_dict:
                route_id_dict[current_voyage_code] = route_id_counter
                route_id_counter += 1
    
        if row[voyage_code] != current_voyage_code:
            stop_id = 1
        else:
            stop_id += 1
    
    if current_row is not None:
        current_row[date] = f"{start_date} to {end_date}"
        current_row['stop_id'] = stop_id
        current_row['route_id'] = route_id_dict[current_voyage_code]
        if start_date == end_date:
            current_row[date] = start_date
        merged_df = pd.concat([merged_df, current_row.to_frame().transpose()], ignore_index=True)
    
    # Update stop_id to range from 1 to the number of rows for each voyage code
    for code in merged_df[voyage_code].unique():
        code_mask = merged_df[voyage_code] == code
        num_rows = code_mask.sum()
        merged_df.loc[code_mask, 'stop_id'] = list(range(1, num_rows + 1))
    
    # Convert stop_id and route_id to integers
    merged_df['stop_id'] = merged_df['stop_id'].astype(int)
    merged_df['route_id'] = merged_df['route_id'].astype(int)
    
    return merged_df
```

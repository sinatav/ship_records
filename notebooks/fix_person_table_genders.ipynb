```python
import pandas as pd
import numpy as np
from tqdm import tqdm
```

```python
df_p = pd.read_csv('person.csv', low_memory=False).rename(columns={'id': 'person_id'})
df_p_id = pd.read_csv('beta_person.csv', low_memory=False).drop('Unnamed: 0', axis=1)
```

```python
df_p_id['child'] = np.nan
df_p['child'] = np.nan
```

```python
for index, row in tqdm(df_p_id.iterrows()):
    if row['gender'] == 'f':
        df_p_id.at[index, 'child'] = 'N'
    elif row['gender'] == 'm':
        df_p_id.at[index, 'child'] = 'N'
    elif row['gender'] == 'fc':
        df_p_id.at[index, 'child'] = 'Y'
        df_p_id.at[index, 'gender'] = 'f'
    elif row['gender'] == 'mc':
        df_p_id.at[index, 'child'] = 'Y'
        df_p_id.at[index, 'gender'] = 'm'
    elif row['gender'] == 'uc':
        df_p_id.at[index, 'child'] = 'Y'
        df_p_id.at[index, 'gender'] = 'u'
    elif row['gender'] == 'c':
        df_p_id.at[index, 'child'] = 'Y'
        df_p_id.at[index, 'gender'] = 'u'
    else:
        df_p_id.at[index, 'child'] = 'u'
        df_p_id.at[index, 'gender'] = 'u'
```

```python
df_p_id.to_csv('beta_person_new')
```

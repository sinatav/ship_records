```python
def duplicate_emails(person: pd.DataFrame) -> pd.DataFrame:
    df = person.drop_duplicates(subset=['email'])
    df2 = pd.concat([df, person])
    df3 = df2.drop_duplicates(keep=False)[['email']].rename(columns={'email':'Email'})
    return df3
```

```python
def calculate_cancellation_rate(group):
    group = group[(group['banned_x'] == 'No') & (group['banned_y'] == 'No')]
    if group.empty:
        return None
    return round((group['status'].isin(['cancelled_by_client', 'cancelled_by_driver']).sum() / len(group)), 2) if len(group) > 0 else 0

def trips_and_users(trips: pd.DataFrame, users: pd.DataFrame) -> pd.DataFrame:
    trips['request_at'] = pd.to_datetime(trips['request_at'])
    if trips.empty or (len([i for i in trips.request_at.between(pd.to_datetime('2013-10-01'), pd.to_datetime('2013-10-03')) if i]) == 0):
        return pd.DataFrame({'Day': [], 'Cancellation Rate': []})
    df = trips.merge(users, left_on='client_id', right_on='users_id').merge(users, left_on='driver_id', right_on='users_id').sort_values(by='request_at')
    start_date = pd.to_datetime('2013-10-01')
    end_date = pd.to_datetime('2013-10-03')
    df = df[(df['request_at'] >= start_date) & (df['request_at'] <= end_date)]

    result_df = df.groupby('request_at').apply(calculate_cancellation_rate)
    result_df.index.name = 'Day'
    
#     return pd.DataFrame(result_df).rename(columns={0:'Cancellation Rate'}).reset_index()
    return result_df
```

```python
def find_first(group):
    return min(group['event_date'])

def game_analysis(activity: pd.DataFrame) -> pd.DataFrame:
    grouped = activity.groupby('player_id').apply(find_first).reset_index()
    return grouped.rename(columns={0:'first_login'})
#     return grouped
```

```python
def calc(group):
    return group[group.event_date == (group.event_date.min() + pd.Timedelta(days=1))]

def gameplay_analysis(activity: pd.DataFrame) -> pd.DataFrame:
    if activity.empty:
        return pd.DataFrame({'fraction': [0]})
    activity['event_date'] = pd.to_datetime(activity['event_date'])
    grouped = activity.groupby('player_id').apply(calc)

    res = [len(grouped)/len(activity['player_id'].unique())]
    return round(pd.DataFrame({'fraction': res}), 2)
```

```python
import re

from io import StringIO
from datetime import datetime


def str_to_date(date_str):
    return datetime.strptime(str(date_str), "%Y%m%d")

def str_to_df(data_str):
    cleaned_data_str = re.sub('[|\-]', '', data_str)
    data_lines = cleaned_data_str.strip().split('\n')
    column_names = [col.strip() for col in data_lines[0].split() if col.strip()]
    data_lines = data_lines[2:]
    df = pd.read_csv(StringIO('\n'.join(data_lines)), delim_whitespace=True, names=column_names)
    
    # Reset the index
    df.reset_index(drop=True, inplace=True)
    
    return df
```

```python
def cal_qua(group):
    avg_ratio = (group['rating'] / group['position']).mean()
    return round(avg_ratio, 2)

def cal_poo(group):
    return round((group['rating'] < 3).mean() * 100, 2)

def queries_stats(queries: pd.DataFrame) -> pd.DataFrame:
    group = queries.groupby('query_name')
    
    quality = group.apply(cal_qua)
    poor_query_percentage = group.apply(cal_poo)
    
    queries = queries.drop_duplicates(subset=['query_name'])

    queries = queries.merge(quality.reset_index()).merge(poor_query_percentage.reset_index(), on='query_name')
    queries = queries.rename(columns={'0_x': 'quality', '0_y': 'poor_query_percentage'})
    queries = queries[['query_name', 'quality', 'poor_query_percentage']]
    
    return queries
```

```python
final_data = pd.read_csv('Downloads/beanutils_cl.csv').drop('Unnamed: 0', axis=1)
final_data = final_data[final_data.FILES == 'PropertyUtils.java']
final_data
```

```python
feature = 'CLASS_COMPLEXITY'

# Check for NaN values in CLASS_COMPLEXITY
nan_values = final_data[feature].isna().sum()
print(f'Number of NaN values in {feature}: {nan_values}')

# If there are NaN values, you can fill them or drop them
final_data[feature] = final_data[feature].fillna(final_data[feature].mean())

# Calculate mean and standard deviation for the 'CLASS_COMPLEXITY' column
mean = final_data[feature].mean()
std_dev = final_data[feature].std()

# Identify outliers with a threshold of 3 standard deviations from the mean
outliers = final_data[(final_data[feature] > mean + 3 * std_dev) | (final_data[feature] < mean - 3 * std_dev)]

# Plotting class_complexity and highlighting outliers using mean and standard deviation method
plt.figure(figsize=(10, 6))
plt.plot(final_data.index, final_data[feature], linestyle='-', label=feature)
plt.scatter(outliers.index, outliers[feature], color='red', label='Outliers')

plt.title(f'{feature} with Outliers Highlighted')
plt.xlabel('Index')
plt.ylabel(feature)
plt.legend()
# plt.grid(True)

# Save the plot as an image
output_image_path = f'{feature}_outliersprp2.png'
plt.savefig(output_image_path)
print(f'Plot saved as {output_image_path}')

# Show the plot
plt.show()
```
